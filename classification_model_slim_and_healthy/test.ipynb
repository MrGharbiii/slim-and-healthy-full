```python
valid_profiles = {
    "metabolique": "metabolique",
    "métabolique": "metabolique",
    "meatboalique": "metabolique",
    "metabolique": "metabolique",
    "Metabolique": "metabolique",
    "métabolique": "metabolique",
    "meatboalique": "metabolique",

    "hormonal": "hormonal",
    "h++": "hormonal",
    "H++": "hormonal",
    "h+": "hormonal",
    "H+": "hormonal",
    "h": "hormonal",
    "H": "hormonal",

    "hormonale": "hormonal",

    "digestif": "digestif",

    "psychologique": "psychologique",
    "psy": "psychologique",
    "psy+": "psychologique",
    "psy++": "psychologique",
    "psy+++": "psychologique",
    "pssy": "psychologique",
    "psy": "psychologique",
    "Psy": "psychologique",
    "psy++": "psychologique",
    "Psy++": "psychologique",
    "psy+++": "psychologique",
    "Psy+++": "psychologique",

    "iatrogene": "iatrogene",
    "iatrogène": "iatrogene",
    "itarogéne": "iatrogene",

    "chrono": "iatrogene",
    "chrno": "iatrogene",
    "chrono+": "iatrogene"
}
def clean_label(label_str):
    # Lowercase and remove +
    raw = str(label_str).lower().replace("+", "")
    # Split by "/" and clean
    parts = [p.strip() for p in raw.split("/")]

    # Map to valid profiles
    cleaned = []
    for p in parts:
        standard = valid_profiles.get(p)
        if standard and standard not in cleaned:
            cleaned.append(standard)

    return sorted(cleaned)  # Alphabetical order

df["LABEL_LIST"] = df["LABEL"].apply(clean_label)
print(df)
```

```python
# Create binary family history features from terrain familial
print("\n=== CREATING BINARY FAMILY HISTORY FEATURES ===")

# Define the family history conditions to extract
family_conditions = [
    'DT1',           # Diabète type 1
    'DT2',           # Diabète type 2  
    'dysthyroidie',  # Dysthyroïdie
    'HTA',           # Hypertension artérielle
    'obésité',       # Obésité (handling variations)
    'dyslipidémie',  # Dyslipidémie (handling variations)
    'Coronaropathie',# Coronaropathie
    'Autisme',       # Autisme
    'polypose familiale', # Polypose familiale
    'Deficit17alphaOH'    # Déficit 17alpha-OH
]

# Create a working copy of X_processed
X_enhanced = X_processed.copy()

# Function to check if a condition is present in terrain familial
def has_family_history(terrain_value, condition):
    if pd.isna(terrain_value) or terrain_value == 'non':
        return 0
    
    terrain_str = str(terrain_value).lower()
    
    # Handle variations in naming
    condition_variations = {
        'obésité': ['obésité', 'obésite'],  # Handle accent variations
        'dyslipidémie': ['dyslipidémie', 'dyslipidemie', 'dyslipidémie/dyslipidémie'],
        'dysthyroidie': ['dysthyroidie', 'dysthyroïdie']
    }
    
    if condition in condition_variations:
        return int(any(var in terrain_str for var in condition_variations[condition]))
    else:
        return int(condition.lower() in terrain_str)

# Create binary features for each family history condition
for condition in family_conditions:
    feature_name = f'fh_{condition.lower().replace(" ", "_")}'
    X_enhanced[feature_name] = X_enhanced['terrain familial'].apply(
        lambda x: has_family_history(x, condition)
    )
    
    # Count positive cases
    positive_count = X_enhanced[feature_name].sum()
    print(f"{feature_name:20s}: {positive_count:3d} patients ({positive_count/len(X_enhanced)*100:.1f}%)")

# Remove the original terrain familial feature
X_enhanced = X_enhanced.drop('terrain familial', axis=1)

print(f"\nOriginal features: {X_processed.shape[1]}")
print(f"Enhanced features: {X_enhanced.shape[1]}")
print(f"Added binary family history features: {len(family_conditions)}")
```

```python
# =============================================================================
# 🚀 MACHINE LEARNING PIPELINE - STEP BY STEP
# =============================================================================

print("🎯 STARTING MACHINE LEARNING PIPELINE")
print("=" * 60)

# Import required libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report, hamming_loss, jaccard_score
from sklearn.metrics import precision_score, recall_score, f1_score
import warnings
warnings.filterwarnings('ignore')

# Check current state
print(f"📊 CURRENT DATA STATE:")
print(f"   Features (X_processed): {X_processed.shape}")
print(f"   Labels (y): {y.shape}")
print(f"   Target classes: {list(y.columns)}")
print(f"   Missing values: {X_processed.isnull().sum().sum()}")
```

```python
# =============================================================================
# STEP 1: CATEGORICAL ENCODING
# =============================================================================

print("\n🔄 STEP 1: CATEGORICAL ENCODING")
print("-" * 40)

# Identify categorical columns
categorical_cols = X_processed.select_dtypes(include=['object']).columns.tolist()
print(f"📝 Categorical columns found: {len(categorical_cols)}")
if categorical_cols:
    print(f"   Columns: {categorical_cols}")
    
    # Apply Label Encoding to categorical columns
    label_encoders = {}
    X_encoded = X_processed.copy()
    
    for col in categorical_cols:
        le = LabelEncoder()
        X_encoded[col] = le.fit_transform(X_processed[col].astype(str))
        label_encoders[col] = le
        print(f"   ✅ Encoded '{col}': {len(le.classes_)} unique values")
        
    print(f"\n✅ Categorical encoding complete!")
else:
    print("   ℹ️ No categorical columns found - all features are already numeric")
    X_encoded = X_processed.copy()

print(f"📊 Final feature matrix: {X_encoded.shape}")
print(f"📊 Data types: {X_encoded.dtypes.value_counts().to_dict()}")
```

```python
# =============================================================================
# STEP 2: TRAIN-TEST SPLIT
# =============================================================================

print("\n🔄 STEP 2: TRAIN-TEST SPLIT")
print("-" * 40)

# Split the data (80% train, 20% test)
# For multi-label, we'll use stratify based on the sum of labels per sample
try:
    # Calculate label distribution for stratification
    y_sum = y.sum(axis=1)  # Number of labels per sample
    
    X_train, X_test, y_train, y_test = train_test_split(
        X_encoded, y, 
        test_size=0.2, 
        random_state=42,
        stratify=y_sum  # Stratify by number of labels per sample
    )
    stratify_used = "by label count"
except:
    # If stratification fails, do regular split
    X_train, X_test, y_train, y_test = train_test_split(
        X_encoded, y,
        test_size=0.2,
        random_state=42
    )
    stratify_used = "random"

print(f"📊 Split strategy: {stratify_used}")
print(f"📊 Training set: {X_train.shape[0]} samples")
print(f"📊 Test set: {X_test.shape[0]} samples")
print(f"📊 Features: {X_train.shape[1]}")

# Show label distribution in train/test
print(f"\n📈 LABEL DISTRIBUTION:")
for col in y.columns:
    train_pos = y_train[col].sum()
    test_pos = y_test[col].sum()
    train_pct = (train_pos / len(y_train)) * 100
    test_pct = (test_pos / len(y_test)) * 100
    print(f"   {col}: Train {train_pos}({train_pct:.1f}%) | Test {test_pos}({test_pct:.1f}%)")
```

```python
# =============================================================================
# STEP 3: MODEL TRAINING
# =============================================================================

print("\n🔄 STEP 3: MODEL TRAINING")
print("-" * 40)

# Initialize Random Forest with MultiOutput wrapper
print("🌳 Training Random Forest Multi-Output Classifier...")

# Base Random Forest parameters
rf_params = {
    'n_estimators': 400,
    'max_depth': 10,
    'min_samples_split': 5,
    'min_samples_leaf': 2,
    'random_state': 42,
    'n_jobs': -1
}

# Create multi-output classifier
rf_base = RandomForestClassifier(**rf_params)
model = MultiOutputClassifier(rf_base, n_jobs=-1)

# Train the model
import time
start_time = time.time()
model.fit(X_train, y_train)
training_time = time.time() - start_time

print(f"✅ Model training completed!")
print(f"⏱️ Training time: {training_time:.2f} seconds")
print(f"🌳 Number of trees: {rf_params['n_estimators']}")
print(f"🎯 Output labels: {len(y.columns)}")
```

```python
# =============================================================================
# STEP 4: PREDICTIONS & EVALUATION
# =============================================================================

print("\n🔄 STEP 4: PREDICTIONS & EVALUATION")
print("-" * 40)

# Make predictions
print("🎯 Making predictions...")
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)

# Calculate overall metrics
print("\n📊 OVERALL PERFORMANCE METRICS:")
print("=" * 50)

# Hamming Loss (lower is better)
hamming = hamming_loss(y_test, y_pred)
print(f"🎯 Hamming Loss: {hamming:.4f}")

# Jaccard Score (higher is better)
jaccard = jaccard_score(y_test, y_pred, average='samples')
print(f"🎯 Jaccard Score: {jaccard:.4f}")

# Macro-averaged metrics
macro_precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
macro_recall = recall_score(y_test, y_pred, average='macro', zero_division=0) 
macro_f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)

print(f"🎯 Macro Precision: {macro_precision:.4f}")
print(f"🎯 Macro Recall: {macro_recall:.4f}")
print(f"🎯 Macro F1-Score: {macro_f1:.4f}")

# Micro-averaged metrics
micro_precision = precision_score(y_test, y_pred, average='micro', zero_division=0)
micro_recall = recall_score(y_test, y_pred, average='micro', zero_division=0)
micro_f1 = f1_score(y_test, y_pred, average='micro', zero_division=0)

print(f"🎯 Micro Precision: {micro_precision:.4f}")
print(f"🎯 Micro Recall: {micro_recall:.4f}")
print(f"🎯 Micro F1-Score: {micro_f1:.4f}")
```

```python
# =============================================================================
# STEP 5: PER-LABEL PERFORMANCE ANALYSIS
# =============================================================================

print("\n🔄 STEP 5: PER-LABEL PERFORMANCE")
print("-" * 40)

# Detailed classification report for each label
print("\n📋 DETAILED CLASSIFICATION REPORT:")
print("=" * 60)

for i, label in enumerate(y.columns):
    print(f"\n🏷️ LABEL: {label}")
    print("-" * 30)
    
    # Get predictions for this label
    y_true_label = y_test.iloc[:, i]
    y_pred_label = y_pred[:, i]
    
    # Calculate metrics for this label
    from sklearn.metrics import precision_recall_fscore_support, accuracy_score
    
    precision, recall, f1, support = precision_recall_fscore_support(
        y_true_label, y_pred_label, average='binary', zero_division=0
    )
    accuracy = accuracy_score(y_true_label, y_pred_label)
    
    # Count predictions
    true_positives = ((y_true_label == 1) & (y_pred_label == 1)).sum()
    false_positives = ((y_true_label == 0) & (y_pred_label == 1)).sum()
    false_negatives = ((y_true_label == 1) & (y_pred_label == 0)).sum()
    true_negatives = ((y_true_label == 0) & (y_pred_label == 0)).sum()
    
    print(f"   Accuracy:  {accuracy:.4f}")
    print(f"   Precision: {precision:.4f}")
    print(f"   Recall:    {recall:.4f}")
    print(f"   F1-Score:  {f1:.4f}")
    print(f"   Support:   {support}")
    print(f"   TP: {true_positives}, FP: {false_positives}, FN: {false_negatives}, TN: {true_negatives}")
    
# Summary table
print("\n📊 PERFORMANCE SUMMARY TABLE:")
print("=" * 70)

results_df = pd.DataFrame({
    'Label': y.columns,
    'Accuracy': [accuracy_score(y_test.iloc[:, i], y_pred[:, i]) for i in range(len(y.columns))],
    'Precision': [precision_score(y_test.iloc[:, i], y_pred[:, i], zero_division=0) for i in range(len(y.columns))],
    'Recall': [recall_score(y_test.iloc[:, i], y_pred[:, i], zero_division=0) for i in range(len(y.columns))],
    'F1-Score': [f1_score(y_test.iloc[:, i], y_pred[:, i], zero_division=0) for i in range(len(y.columns))],
    'Support': [y_test.iloc[:, i].sum() for i in range(len(y.columns))]
})

print(results_df.round(4))
```

```python
# =============================================================================
# STEP 6: FEATURE IMPORTANCE ANALYSIS
# =============================================================================

print("\n🔄 STEP 6: FEATURE IMPORTANCE ANALYSIS")
print("-" * 40)

# Get feature importance from each classifier
print("🌟 Analyzing feature importance...")

# Average feature importance across all classifiers
feature_importance_list = []
for i, estimator in enumerate(model.estimators_):
    importance = estimator.feature_importances_
    feature_importance_list.append(importance)
    print(f"   ✅ Extracted importance from classifier {i+1} ({y.columns[i]})")

# Calculate average importance
avg_importance = np.mean(feature_importance_list, axis=0)

# Create feature importance dataframe
feature_importance_df = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': avg_importance
}).sort_values('Importance', ascending=False)

print(f"\n🏆 TOP 20 MOST IMPORTANT FEATURES:")
print("=" * 50)
print(feature_importance_df.head(20).to_string(index=False))

# Feature importance by category
print(f"\n📋 FEATURE IMPORTANCE BY CATEGORY:")
print("=" * 40)

# Group features by category (based on naming patterns)
categories = {
    'Demographics': ['age', 'sexe', 'taille', 'poids', 'IMC'],
    'Family_History': [col for col in X_encoded.columns if col.startswith('famille_')],
    'Psychology': [col for col in X_encoded.columns if col.startswith('psy_')],
    'Treatments': [col for col in X_encoded.columns if col.startswith('tt_')],
    'Medical_Treatments': [col for col in X_encoded.columns if col.startswith('med_')],
    'Laboratory': ['TSH', 'glycemie', 'cholesterol', 'triglycerides'],
    'Other': []
}

# Add remaining features to 'Other'
all_categorized = []
for cat_features in categories.values():
    all_categorized.extend(cat_features)

categories['Other'] = [col for col in X_encoded.columns if col not in all_categorized]

# Calculate average importance by category
for category, features in categories.items():
    if features:
        category_features = [f for f in features if f in feature_importance_df['Feature'].values]
        if category_features:
            avg_cat_importance = feature_importance_df[feature_importance_df['Feature'].isin(category_features)]['Importance'].mean()
            print(f"   {category:20s}: {avg_cat_importance:.6f} (n={len(category_features)})")
```

```python
# =============================================================================
# 🎉 PIPELINE COMPLETION SUMMARY
# =============================================================================

print("\n" + "=" * 70)
print("🎉 MACHINE LEARNING PIPELINE COMPLETED SUCCESSFULLY!")
print("=" * 70)

print(f"\n📊 FINAL SUMMARY:")
print(f"   📋 Dataset: {X_encoded.shape[0]} patients, {X_encoded.shape[1]} features")
print(f"   🎯 Target labels: {len(y.columns)} obesity profiles")
print(f"   🚂 Training samples: {X_train.shape[0]}")
print(f"   🧪 Test samples: {X_test.shape[0]}")
print(f"   ⏱️ Training time: {training_time:.2f} seconds")

print(f"\n🏆 BEST PERFORMANCE METRICS:")
print(f"   🎯 Overall Jaccard Score: {jaccard:.4f}")
print(f"   🎯 Macro F1-Score: {macro_f1:.4f}")
print(f"   🎯 Micro F1-Score: {micro_f1:.4f}")
print(f"   🎯 Hamming Loss: {hamming:.4f}")

print(f"\n🏷️ PER-LABEL PERFORMANCE:")
for i, label in enumerate(y.columns):
    label_f1 = f1_score(y_test.iloc[:, i], y_pred[:, i], zero_division=0)
    label_acc = accuracy_score(y_test.iloc[:, i], y_pred[:, i])
    print(f"   {label:15s}: F1={label_f1:.3f}, Acc={label_acc:.3f}")

print(f"\n🌟 TOP 3 MOST IMPORTANT FEATURES:")
for i, (_, row) in enumerate(feature_importance_df.head(3).iterrows()):
    print(f"   {i+1}. {row['Feature']}: {row['Importance']:.6f}")

print(f"\n✅ All steps completed successfully!")
print(f"✅ Model is ready for predictions on new patient data")
print("=" * 70)
```

```python
# Complete test pipeline for new patient prediction

# Test patient data
new_patient_data = {
    "Sexe": "M",
    "profession": "etudiant",
    "Nb enfants": 0,
    "DT2": "non",
    "DT1": "non",
    "SAS": "non",
    "Hypothyroidie": "non",
    "SOPK": "non",
    "contraception hormonale": "non",
    "accouchement / avortement< 2 ans": "non",
    "Menopause/Peri": "non",
    "troubles digestif": "non",
    "trouble du sommeil": "non",
    "niveau de stress": "élevé",
    "Psychothérapie": "non",
    "traitement anti-obésité antérieur": "non",
    "activité": "sédentaire",
    "régimes antérieures": "non",
    "Tabac": "non",
    "alcool": "non",
    "Nombre de séance sport/semaine": 0,
    "Travail posté": "oui",
    "Taille": 178,
    "P0": 68,
    "MG0": 25,
    "MM0": 42.7,
    "OBJECTIF MG": 6,
    "Objectif MM": 4,
    "TSH": 4,
    "Age": 24,
    "terrain familial": "non",
    "trouble psy": "non",
    "traitements": "non",
    "TT medical": "non"
}

print("🎯 TESTING NEW PATIENT PREDICTION")
print("=" * 50)

print("Test patient profile:")
for key, value in new_patient_data.items():
    print(f"  {key}: {value}")

# Convert to DataFrame
test_df = pd.DataFrame([new_patient_data])

# Apply the SAME preprocessing pipeline as training data
print("\n🔄 Applying preprocessing pipeline...")

# 1. Handle terrain familial -> create family history features
family_conditions = ['DT1', 'DT2', 'dysthyroidie', 'HTA', 'obésité', 'dyslipidémie', 'Coronaropathie', 'Autisme', 'polypose familiale', 'Deficit17alphaOH']

def has_family_history(terrain_value, condition):
    if pd.isna(terrain_value) or terrain_value == 'non':
        return 0
    terrain_str = str(terrain_value).lower()
    condition_variations = {
        'obésité': ['obésité', 'obésite'],
        'dyslipidémie': ['dyslipidémie', 'dyslipidemie'],
        'dysthyroidie': ['dysthyroidie', 'dysthyroïdie']
    }
    if condition in condition_variations:
        return int(any(var in terrain_str for var in condition_variations[condition]))
    else:
        return int(condition.lower() in terrain_str)

for condition in family_conditions:
    feature_name = f'fh_{condition.lower().replace(" ", "_")}'
    test_df[feature_name] = test_df['terrain familial'].apply(lambda x: has_family_history(x, condition))

test_df = test_df.drop('terrain familial', axis=1)

# 2. Handle trouble psy -> create psychological features
psy_conditions = ['dépression', 'TCA', 'troubles anxieux', 'anxiété', 'troubles du sommeil', 'bipolarité', 'TOC', 'trouble panique', 'stress post-traumatique', 'troubles de l humeur']

def has_psy_condition(psy_value, condition):
    if pd.isna(psy_value) or psy_value == 'non':
        return 0
    psy_str = str(psy_value).lower().replace('/', ' ').replace('+', ' ')
    condition_variations = {
        'troubles anxieux': ['troubles anxieux', 'anxieux', 'trouble anxieux'],
        'TCA': ['tca', 'troubles du comportement alimentaire'],
        'dépression': ['dépression', 'depression'],
        'troubles du sommeil': ['troubles du sommeil', 'trouble du sommeil'],
        'anxiété': ['anxiété', 'anxiete', 'angoisse'],
        'bipolarité': ['bipolarité', 'bipolaire']
    }
    if condition in condition_variations:
        return int(any(var in psy_str for var in condition_variations[condition]))
    else:
        return int(condition.lower() in psy_str)

for condition in psy_conditions:
    clean_condition = condition.lower().replace(" ", "_").replace("'", "")
    feature_name = f'psy_{clean_condition}'
    test_df[feature_name] = test_df['trouble psy'].apply(lambda x: has_psy_condition(x, condition))

test_df = test_df.drop('trouble psy', axis=1)

# 3. Handle treatments -> create treatment features
treatment_conditions = ['anti-dépressif', 'levothyrox', 'hypolipémiant', 'anti-hypertenseur', 'Metformine', 'anxiolytique', 'insuline', 'sulfamide', 'contraceptif', 'corticoides', 'FIV', 'fer', 'statine', 'antalgique', 'anti-allergique']

def has_treatment(treatment_value, condition):
    if pd.isna(treatment_value) or treatment_value == 'non':
        return 0 
    treatment_str = str(treatment_value).lower().replace('/', ' ').replace('+', ' ')
    return int(condition.lower() in treatment_str)

for condition in treatment_conditions:
    clean_condition = condition.lower().replace("-", "_").replace("é", "e").replace("è", "e")
    feature_name = f'treat_{clean_condition}'
    test_df[feature_name] = test_df['traitements'].apply(lambda x: has_treatment(x, condition))

test_df = test_df.drop('traitements', axis=1)

# 4. Handle medical treatments
medical_conditions = ['Levothyrox', 'fer', 'Probiotique', 'Metformine', 'IDPP4', 'ISGLT2', 'Ozempic', 'cabergoline', 'Mounjaro', 'Victoza', 'contraceptif', 'antidiabétique', 'testosterone', 'SAS', 'fibrates', 'antidépresseur', 'corticothérapie']

def has_medical_treatment(medical_value, condition):
    if pd.isna(medical_value) or medical_value == 'non':
        return 0
    medical_str = str(medical_value).lower().replace('/', ' ').replace('+', ' ')
    return int(condition.lower() in medical_str)

for condition in medical_conditions:
    clean_condition = condition.lower().replace("é", "e").replace("è", "e").replace("ê", "e")
    feature_name = f'med_{clean_condition}'
    test_df[feature_name] = test_df['TT medical'].apply(lambda x: has_medical_treatment(x, condition))

test_df = test_df.drop('TT medical', axis=1)

# 5. Ensure all features match training data
print(f"Training features: {len(X_encoded.columns)}")
print(f"Test features before alignment: {len(test_df.columns)}")

# Add missing features with default values (0 for binary, appropriate defaults for others)
for col in X_encoded.columns:
    if col not in test_df.columns:
        test_df[col] = 0  # Default value for missing binary features

# Reorder columns to match training data
test_df = test_df[X_encoded.columns]

print(f"Test features after alignment: {len(test_df.columns)}")

# 6. Apply categorical encoding using the same encoders from training
test_processed = test_df.copy()

for col in label_encoders.keys():
    if col in test_processed.columns:
        # Handle new categories not seen during training
        test_value = str(test_processed[col].iloc[0])
        if test_value in label_encoders[col].classes_:
            test_processed[col] = label_encoders[col].transform([test_value])
        else:
            # Assign to most frequent class or 0
            test_processed[col] = 0
            print(f"⚠️ New category '{test_value}' in {col}, assigned default value")

print("✅ Preprocessing completed!")

# 7. Make prediction
print("\n🎯 Making prediction...")
prediction = model.predict(test_processed)
prediction_proba = model.predict_proba(test_processed)

print("\n📊 PREDICTION RESULTS:")
print("=" * 40)

predicted_profiles = []
for i in range(len(y.columns)):
    if prediction[0][i] == 1:
        predicted_profiles.append(y.columns[i])
        prob = prediction_proba[i][0][1] if len(prediction_proba[i][0]) > 1 else prediction_proba[i][0][0]
        print(f"✅ {y.columns[i]}: POSITIVE (probability: {prob:.3f})")
    else:
        prob = prediction_proba[i][0][0] if len(prediction_proba[i][0]) > 1 else 1 - prediction_proba[i][0][0]
        print(f"❌ {y.columns[i]}: NEGATIVE (probability: {prob:.3f})")

# Updated prediction results display with percentages for all labels
print("\n📊 PREDICTION RESULTS - PROBABILITY PERCENTAGES:")
print("=" * 55)

# Display percentages for all labels
predicted_profiles_new = []
label_probabilities = []

for i in range(len(y.columns)):
    # Get the probability for positive class (class 1)
    pos_prob = prediction_proba[i][0][1] if len(prediction_proba[i][0]) > 1 else prediction_proba[i][0][0]
    
    # Convert to percentage
    percentage = pos_prob * 100
    label_probabilities.append((y.columns[i], percentage))
    
    # Check if predicted positive (above 50%)
    if prediction[0][i] == 1:
        predicted_profiles_new.append(y.columns[i])
        print(f"🟢 {y.columns[i]:<15}: {percentage:5.1f}% (POSITIVE)")
    else:
        print(f"🔴 {y.columns[i]:<15}: {percentage:5.1f}% (NEGATIVE)")

# Sort by probability and show ranking
print("\n📊 RANKING BY PROBABILITY:")
print("-" * 35)
sorted_probs = sorted(label_probabilities, key=lambda x: x[1], reverse=True)
for rank, (label, prob) in enumerate(sorted_probs, 1):
    print(f"{rank}. {label:<15}: {prob:5.1f}%")

if predicted_profiles_new:
    print(f"\n🎯 PREDICTED PROFILES: {', '.join(predicted_profiles_new)}")
else:
    print(f"\n🎯 NO SPECIFIC PROFILES PREDICTED")

print("\n✅ Prediction completed successfully!")
```
